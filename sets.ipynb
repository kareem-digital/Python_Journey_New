##### Creating Sets

```python
# empty
s = set()
print(s)
print(type(s))
# 1D and 2D
s1 = {1,2,3}
print(s1)
# s2 = {1,2,3,{4,5}}
# print(s2)
# homo and hetro
s3 = {1,'hello',4.5,(1,2,3)}
print(s3)
# using type conversion

s4 = set([1,2,3])
print(s4)
# duplicates not allowed
s5 = {1,1,2,2,3,3}
print(s5)
# set can't have mutable items
# s6 = {1,2,[3,4]}
# print(s6)
s6 = {1 , 'hello', 4.5 , True}
print(s6)
```

##### Set Operation

```python
# union() or |
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}
# s1 | s2
# s1.union(s2)

# intersection() or &
# s1 & s2 
# s1.intersection(s2)
# s1.intersection_update(s2)
# print(s1)
# s1 - s2 
# s1.difference(s2)
# s1 ^ s2 
# s1.symmetric_difference(s2)
1 in s1 
1 not in s1 
for i in s1:
  print(i)
```

```python
x = {"apple", "banana", "cherry"}
y = {"google", "microsoft", "apple"}

x.intersection_update(y)

print(x)
```

##### difference between intersection and intersection_update in python

```python
cities = {"Tokyo", "Madrid", "Berlin", "Delhi"}
cities2 = {"Tokyo", "Seoul", "Kabul", "Madrid"}

# cities3 = cities.intersection(cities2)
cities3 = cities & cities2

print(f"Original set 'cities': {cities}")
print(f"New set 'cities3' (intersection): {cities3}")
```

```python
cities = {"Tokyo", "Madrid", "Berlin", "Delhi"}
cities2 = {"Tokyo", "Seoul", "Kabul", "Madrid"}

# cities.intersection_update(cities2)
cities &= cities2

print(f"Updated set 'cities': {cities}")
print(f"Set 'cities2' (unchanged): {cities2}")
```

```python
# isdisjoint/issubset/issuperset
s1 = {1,2,3,4}
s2 = {7,8,5,6}

s1.issuperset(s2)
```

```python
x = {"a", "b", "c"}
y = {"f", "e", "d", "c", "b", "a"}

z = x.issubset(y)

print(z)
```

```python
x = {"f", "e", "d", "c", "b", "a"}
y = {"a", "b", "c"}

z = x.issuperset(y)

print(z)
```

```python
x = {"f", "e", "d", "c", "b"}
y = {"a", "b", "c"}

z = x.issuperset(y)

print(z)
```
