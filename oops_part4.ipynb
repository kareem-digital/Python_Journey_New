````python
# Method overriding
class Shape:
    def area(self):
        print("Calculating area...")

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):                         # Overriding!
        print(f"Circle Area = {3.14 * self.radius * self.radius}")

class Rectangle(Shape):
    def __init__(self, length, breadth):
        self.length = length
        self.breadth = breadth

    def area(self):                         # Overriding!
        print(f"Rectangle Area = {self.length * self.breadth}")

class Triangle(Shape):
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):                         # Overriding!
        print(f"Triangle Area = {0.5 * self.base * self.height}")

shapes = [Circle(7), Rectangle(4, 5), Triangle(6, 8)]

for shape in shapes:
    shape.area()            # Same call → Different formula!


````

````python
# ✅ WITH Overriding — clean, scalable, professional
class Payment:
    def __init__(self, amount):
        self.amount = amount

    def process(self):                          # Generic method
        print("Processing payment...")

class CreditCard(Payment):
    def process(self):                          # Overriding!
        print(f"Processing Credit Card of ₹{self.amount}")
        print("Checking card limit...")
        print("Sending OTP...")

class UPI(Payment):
    def process(self):                          # Overriding!
        print(f"Processing UPI of ₹{self.amount}")
        print("Checking UPI ID...")

class NetBanking(Payment):
    def process(self):                          # Overriding!
        print(f"Processing NetBanking of ₹{self.amount}")
        print("Redirecting to bank...")

# ONE LINE handles ALL payment types!
payments = [CreditCard(5000), UPI(2000), NetBanking(10000)]

for payment in payments:
    payment.process()           # Same call → Different behavior ✅
    print("---")

# Processing Credit Card of ₹5000
# Checking card limit...
# Sending OTP...
# ---
# Processing UPI of ₹2000
# Checking UPI ID...
# ---
# Processing NetBanking of ₹10000
# Redirecting to bank...
# ---
````

````python
# Method Overloading
# ❌ This does NOT work in Python!
class Calculator:
    def add(self, a, b):
        return a + b

    def add(self, a, b, c):       # This REPLACES the above!
        return a + b + c

c = Calculator()
# print(c.add(2, 3))        # ❌ TypeError! (needs 3 args now)
print(c.add(2, 3, 4))     # ✅ 9
````

````python
# ✅ Python's Way — Default Arguments
class Calculator:
    def add(self, a, b, c=0, d=0):    # c and d are optional
        return a + b + c + d

calc = Calculator()

print(calc.add(2, 3))           # ✅ 5   (c=0, d=0 by default)
print(calc.add(2, 3, 4))        # ✅ 9   (d=0 by default)
print(calc.add(2, 3, 4, 5))     # ✅ 14  (all provided)
````

````python
# ✅ Python's Way — *args (Most Flexible)
class Calculator:
    def add(self, *args):             # *args = any number of arguments
        total = 0
        for num in args:
            total += num
        return total

calc = Calculator()

print(calc.add(2, 3))               # ✅ 5
print(calc.add(2, 3, 4))            # ✅ 9
print(calc.add(2, 3, 4, 5))         # ✅ 14
print(calc.add(10, 20, 30, 40, 50)) # ✅ 150
````

````python
class Student:
    def introduce(self, name, age=None, city=None):
        print(f"Name: {name}")
        if age:
            print(f"Age: {age}")
        if city:
            print(f"City: {city}")
        print("---")

s = Student()
s.introduce("Rahul")                    # Only name
s.introduce("Priya", 20)               # Name + age
s.introduce("Amit", 22, "Mumbai")      # Name + age + city
````

````python
# Operator Overloading
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def __add__(self, other):               # Overloading + operator
        total = self.marks + other.marks
        return f"Combined Marks = {total}"

s1 = Student("Rahul", 80)
s2 = Student("Priya", 90)

print(s1 + s2)  

"""
**What Python does behind the scenes:**
```
s1 + s2
  ↓
Python sees + operator on custom object
  ↓
Calls s1.__add__(s2) automatically
  ↓
Returns "Combined Marks = 170"
"""
````

````python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def __add__(self, other):               # + operator
        return self.marks + other.marks

    def __gt__(self, other):                # > operator
        return self.marks > other.marks

    def __lt__(self, other):                # < operator
        return self.marks < other.marks

    def __eq__(self, other):                # == operator
        return self.marks == other.marks

    def __str__(self):                      # print() support
        return f"Student({self.name}, Marks={self.marks})"

s1 = Student("Rahul", 80)
s2 = Student("Priya", 90)
s3 = Student("Amit",  80)

print(s1 + s2)      # ✅ 170       (adds marks)
print(s1 > s2)      # ✅ False     (80 > 90?)
print(s1 < s2)      # ✅ True      (80 < 90?)
print(s1 == s3)     # ✅ True      (80 == 80?)
print(s1)           # ✅ Student(Rahul, Marks=80)
````

````python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def __add__(self, other):               # Add two products
        return f"Total Price = ₹{self.price + other.price}"

    def __gt__(self, other):                # Which is expensive?
        if self.price > other.price:
            return f"{self.name} is more expensive"
        else:
            return f"{other.name} is more expensive"

    def __str__(self):
        return f"{self.name} costs ₹{self.price}"

p1 = Product("Phone", 50000)
p2 = Product("Laptop", 80000)

print(p1)           # ✅ Phone costs ₹50000
print(p2)           # ✅ Laptop costs ₹80000
print(p1 + p2)      # ✅ Total Price = ₹130000
print(p1 > p2)      # ✅ Laptop is more expensive



````

````python
# ✅ WITH Operator Overloading — reads like plain English!
class Cart:
    def __init__(self, items, total):
        self.items = items
        self.total = total

    def __add__(self, other):                   # + operator
        combined_items = self.items + other.items
        combined_total = self.total + other.total
        return Cart(combined_items, combined_total)

    def __gt__(self, other):                    # > operator
        return self.total > other.total

    def __lt__(self, other):                    # < operator
        return self.total < other.total

    def __str__(self):                          # print() support
        return f"Cart: {self.items} | Total: ₹{self.total}"

c1 = Cart(["Phone", "Case"], 55000)
c2 = Cart(["Laptop"], 80000)

combined = c1 + c2                  # ✅ Natural!
print(combined)                     # ✅ Cart: ['Phone', 'Case', 'Laptop'] | Total: ₹135000

print(c1 > c2)                      # ✅ False  (is c1 more expensive?)
print(c1 < c2) 
````

````python
# ❌ WITHOUT Operator Overloading — ugly method calls
class Cart:
    def __init__(self, items, total):
        self.items = items
        self.total = total

    def combine(self, other):
        return self.total + other.total

    def is_more_expensive(self, other):
        return self.total > other.total

c1 = Cart(["Phone", "Case"], 55000)
c2 = Cart(["Laptop"], 80000)

print(c1.combine(c2))               # 😫 135000 (unnatural)
print(c1.is_more_expensive(c2))     # 😫 False (hard to read)
````
