```python
# cartesian products -> List comprehension on 2 lists together
L1 = [1,2,3,4]
L2 = [5,6,7,8]

cartesian_prod = [i * j for i in L1 for j in L2]
cartesian_prod
```

```python
# Write a program to replace an item with a different item if found in the list 
L = [1,2,3,4,5,3]
# replace 3 with 300
item_to_find = 3
new_item = 300

new_L = [new_item if x == item_to_find else x for x in L]

print(f"Original list L: {L}")
print(f"New list new_L: {new_L}")
```

```python
# The original list
L = [1, 2, 3, 4, 5, 3]
print(f"Original list: {L}")

# The item to find and replace
item_to_find = 3

# The new item to use as a replacement
new_item = 300

# Iterate through the list by index and replace the item if found
for i in range(len(L)):
    if L[i] == item_to_find:
        L[i] = new_item

# The modified list
print(f"Modified list: {L}")
```

```python
# Write a program that can convert a 2D list to 1D list
list_2d = [[1, 2, 3], [4, 5], [6, 7, 8, 9]]

list_1d = [item for sublist in list_2d for item in sublist]

print(list_1d)
# Output: [1, 2, 3, 4, 5, 6, 7, 8, 9]
```

```python
list_2d = [[1, 2, 3], [4, 5], [6, 7, 8, 9]]
list_1d = []

for sublist in list_2d:
    for item in sublist:
        list_1d.append(item)

print(list_1d)
# Output: [1, 2, 3, 4, 5, 6, 7, 8, 9]
```

```python
# Write a program to remove duplicate items from a list

L = [1,2,1,2,3,4,5,3,4]
print(f"Original list: {L}")
unique_L = []

for item in L:
    if item not in unique_L:
        unique_L.append(item)

print(f"Unique list: {unique_L}")
```

```python
# Find the greatest element and print its index too?
l = [12 , 435 , 67 , 89 , 23 , 25, 690] 
largest =  l[0]
index = 0
for i in range(len(l)):
    if l[i] > largest:
        largest = l[i]
        index = i

print(f"Largest Element: {largest}")
print(f"Largest Element Index Position: {index}")

```

```python
# Find the second greatest element?
l = [12 ,  67 , 89 , 23 , 25, 690, 190] 
largest = l[0]
sec_largest = l[0] 

for i in l:
    if i > largest: 
        sec_largest = largest 
        largest = i 
    elif i > sec_largest:
        sec_largest = i

print(f"Largest: {largest}")
print(f"Second Largest: {sec_largest}")
```
