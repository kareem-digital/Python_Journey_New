##### Iterators and Iterables in Python

```python
# Iteration
"""
Iteration is a general term for taking each item of something, one after another. Any time you use a loop
,explicit or implicit, to go over a grp of items, that is iteration
"""
# Examples
numbers = [1 , 2 , 3 , 4]
for num in numbers:
    print(num)
```

```python
# What is Iterator
"""
An iterator is an object that allows the programmer to traverse through a sequence of data without having
to store the entire data in the memory
"""
# Example
# Problem without iterator
L = [x for x in range(1 , 10000)]
# for i in L:
#     print(i * 2)

import sys 
print(sys.getsizeof(L)/64)

# same problem with range 
x = range(1 , 10000000)
# for i in x:
#     print(i * 2)
print(sys.getsizeof(x)/64)
```

```python
# What is Iterable 
"""
Iterable is an object, which one can iterate over
It generates an Iterator when passed to iter() method
List, range , dict , str , set , tuple all are iterable where we can loop over to get all the items one by one
"""
# Example
L = [1 , 2 , 3]
type(L)
# L is an iterable
my_iter = iter(L)
type(my_iter)
```

```python
# Points to remember
# -> Every Iterator is also an iterable (You can loop over on iterator as well)
# -> Not all iterables are iterators (List is an iterable but it is not an iterator bcz in List we store the full item in memory)

# Trick
# -> Every iterable has an iter function
# -> Every iterator has both iter function as well as next function
```

##### making our own for loop

```python
def mera_khudka_for_loop(iterable):
    iterator = iter(iterable)

    while True:
        try:
            print(next(iterator))
        except StopIteration:
            break
```

```python
a = [1 , 2 , 3]
b = range(1 , 11)
c = (10 , 20 , 30)
d = {40 , 50 , 60}
e = {0: 1 , 1 : 1 , 2 : 2}
mera_khudka_for_loop(a)
print("="*60)
mera_khudka_for_loop(b)
print("="*60)
mera_khudka_for_loop(c)
print("="*60)
mera_khudka_for_loop(d)
print("="*60)
mera_khudka_for_loop(e)
```

```python
print("="*70)
print("Understanding iter() Behavior")
print("="*70)

# Step 1: Create iterable (list)
num = [1, 2, 3]
print(f"\n1️⃣ Original list (iterable):")
print(f"   num = {num}")
print(f"   ID: {id(num)}")
print(f"   Type: {type(num)}")

# Step 2: Convert iterable to iterator
iter_obj = iter(num)
print(f"\n2️⃣ First iterator (iter_obj):")
print(f"   ID: {id(iter_obj)}")
print(f"   Type: {type(iter_obj)}")

# Step 3: Call iter() on iterator AGAIN
iter_obj2 = iter(iter_obj)
print(f"\n3️⃣ Second iterator (iter_obj2):")
print(f"   ID: {id(iter_obj2)}")
print(f"   Type: {type(iter_obj2)}")

# Step 4: Check if they're the same object
print(f"\n4️⃣ Comparison:")
print(f"   Are iter_obj and iter_obj2 the SAME object?")
print(f"   iter_obj is iter_obj2: {iter_obj is iter_obj2} ✅")
print(f"   Same memory address? {id(iter_obj) == id(iter_obj2)} ✅")

# Step 5: Prove they share the same state
print(f"\n5️⃣ They share the same state:")
print(f"   next(iter_obj): {next(iter_obj)}")
print(f"   next(iter_obj2): {next(iter_obj2)}")  # Continues from where iter_obj left off!
print(f"   next(iter_obj): {next(iter_obj)}")
print("   👆 All three calls move the SAME iterator forward!")
```

##### Lets create our own range() function

```python
class MeraRange:
    def __init__(self, start , end):
        self.start = start
        self.end = end

    def __iter__(self):
        return MeraRangeKaIterator(self)
     
```

```python
class MeraRangeKaIterator:
    def __init__(self , iterable_obj):
        self.iterable = iterable_obj

    def __iter__(self):
        return self

    def __next__(self):
        if self.iterable.start >= self.iterable.end:
            raise StopIteration
        
        current = self.iterable.start
        self.iterable.start += 1
        return current
```

```python
# ==================== ITERABLE CLASS ====================
class MeraRange:
    """
    This is the ITERABLE class (like list, tuple)
    It's just a container that stores the range parameters
    """
    
    def __init__(self, start, end):
        """
        Constructor: Initialize with start and end values
        """
        self.start = start  # Store starting number
        self.end = end      # Store ending number (exclusive)
        print(f"  🏗️ MeraRange created: start={start}, end={end}")
    
    def __iter__(self):
        """
        This makes MeraRange ITERABLE
        When someone does: for x in MeraRange(1, 5)
        Python calls this method
        
        Returns: A NEW iterator object
        """
        print(f"  🔄 __iter__() called on MeraRange")
        print(f"     Creating NEW iterator from {self.start} to {self.end}")
        return MeraRangeKaIterator(self.start, self.end)


# ==================== ITERATOR CLASS ====================
class MeraRangeKaIterator:
    """
    This is the ITERATOR class
    It does the actual work of moving through numbers
    """
    
    def __init__(self, start, end):
        """
        Constructor: Initialize iterator's state
        """
        self.current = start  # Track current position
        self.end = end        # Know when to stop
        print(f"    🎯 Iterator initialized: current={start}, end={end}")
    
    def __iter__(self):
        """
        This makes the iterator work in for loops
        Returns itself (this is the rule for iterators!)
        """
        print(f"    🔁 __iter__() called on Iterator (returning self)")
        return self  # Return itself
    
    def __next__(self):
        """
        This method gets called each time we need the next value
        Returns: Next number in sequence
        Raises: StopIteration when done
        """
        print(f"    ➡️ __next__() called: current={self.current}, end={self.end}")
        
        # Check if we've reached the end
        if self.current >= self.end:
            print(f"    🛑 Reached end! Raising StopIteration")
            raise StopIteration
        
        # Get current value
        value = self.current
        print(f"    ✅ Returning: {value}")
        
        # Move to next position
        self.current += 1
        print(f"    📍 Moving current to: {self.current}")
        
        return value


# ==================== USING THE CODE ====================
print("="*70)
print("EXECUTION STARTS")
print("="*70)

for i in MeraRange(1, 5):
    print(f"  📦 Got value in for loop: {i}\n")

print("="*70)
print("EXECUTION COMPLETE")
print("="*70)
```
