##### Abstraction in OOPs

```python
from abc import ABC , abstractmethod
class Shape(ABC):  # Abstract Class
    
    @abstractmethod
    def area(self):   # Abstract Method
        pass
```

```python
class Circle(Shape):
    
    def __init__(self, radius):
        self.radius = radius
        
    def area(self):
        return 3.14 * self.radius * self.radius
```

```python
c = Circle(5)
print(c.area())
```

```python
# Example 1: Without Abstraction (Not Recommended)
class Animal:
    def make_sound(self):
        pass  # What sound? Not clear!

class Dog(Animal):
    def make_sound(self):
        print("Bark!")

class Cat(Animal):
    def make_sound(self):
        pass  # Forgot to implement! No error!

# Problem: Nothing forces Cat to implement make_sound(). This can cause bugs.
```

```python
# Example 2: With Abstraction (Recommended)
from abc import ABC, abstractmethod

# Abstract class - cannot create objects directly
class Animal(ABC):
    
    @abstractmethod
    def make_sound(self):
        pass  # Must be implemented by child classes
    
    @abstractmethod
    def move(self):
        pass # Must be implemented by child classes
```

```python
# Concrete class - provides implementation
class Dog(Animal):
    def make_sound(self):
        print("Bark! Bark!")
    
    def move(self):
        print("Running on four legs")
```

```python
# Using the classes
dog = Dog()
dog.make_sound()  
dog.move()  
```

```python
class Bird(Animal):
    def make_sound(self):
        print("Chirp! Chirp!")
    
    def move(self):
        print("Flying in the sky")
```

```python
bird = Bird()
bird.make_sound()
bird.move()
```

```python
# What if we forget to implement abstract methods?
class Cat(Animal):
    def make_sound(self):
        print("Meow!")
    # Forgot to implement move()!

# This will cause an ERROR:
# cat = Cat()  
# TypeError: Can't instantiate abstract class Cat 
# with abstract method move

# Python forces you to implement all abstract methods!
```

```python
# Practical Example: Payment System
from abc import ABC, abstractmethod

# Abstract Payment class
class Payment(ABC):
    
    @abstractmethod
    def pay(self, amount):
        pass
    
    @abstractmethod
    def refund(self, amount):
        pass
```

```python
# Different payment methods
class CreditCard(Payment):
    def pay(self, amount):
        print(f"Paid ${amount} using Credit Card")
    
    def refund(self, amount):
        print(f"Refunded ${amount} to Credit Card")

class PayPal(Payment):
    def pay(self, amount):
        print(f"Paid ${amount} using PayPal")
    
    def refund(self, amount):
        print(f"Refunded ${amount} to PayPal")

class Bitcoin(Payment):
    def pay(self, amount):
        print(f"Paid ${amount} using Bitcoin")
    
    def refund(self, amount):
        print(f"Refunded ${amount} to Bitcoin wallet")
```

```python
# Using abstraction
def process_payment(payment_method, amount):
    payment_method.pay(amount)
```

```python
# All payment methods work with the same interface
card = CreditCard()
paypal = PayPal()
crypto = Bitcoin()
```

```python
process_payment(card, 100)    
process_payment(paypal, 50)   
process_payment(crypto, 200)  
```

```python
# Complete Real-Time Implementation
from abc import ABC, abstractmethod

# Abstract Base Class - Blueprint for all banks
class BankApp(ABC):
    
    # Common method - all banks use same database
    def database(self):
        print('✅ Connected to central database')
    
    # Common method - all banks have customer login
    def customer_login(self, username):
        print(f'👤 Welcome {username}!')
        self.database()
    
    # Abstract methods - each bank implements differently
    @abstractmethod
    def security(self):
        pass
    
    @abstractmethod
    def calculate_interest(self, amount, years):
        pass
    
    @abstractmethod
    def withdrawal_limit(self):
        pass
```

```python
# ==================== HDFC BANK ====================
class HDFCBank(BankApp):
    
    def security(self):
        print('🔐 HDFC Security: OTP + Fingerprint + Face Recognition')
    
    def calculate_interest(self, amount, years):
        # HDFC gives 7% interest
        interest = amount * 0.07 * years
        total = amount + interest
        print(f'💰 HDFC Interest (7%): ₹{interest}')
        print(f'💵 Total Amount: ₹{total}')
        return total
    
    def withdrawal_limit(self):
        print('🏧 HDFC Daily Withdrawal Limit: ₹50,000')
```

```python
# ==================== MAIN APPLICATION ====================
print("="*60)
print("🏦 MULTI-BANK APPLICATION")
print("="*60)

# Customer chooses HDFC Bank
print("\n📱 Customer 1: Using HDFC Bank")
print("-"*60)
hdfc_customer = HDFCBank()
hdfc_customer.customer_login('Rajesh')
hdfc_customer.security()
hdfc_customer.calculate_interest(100000, 2)
hdfc_customer.withdrawal_limit()
```

```python
# ==================== SBI BANK ====================
class SBIBank(BankApp):
    
    def security(self):
        print('🔐 SBI Security: OTP + Security Questions')
    
    def calculate_interest(self, amount, years):
        # SBI gives 6.5% interest
        interest = amount * 0.065 * years
        total = amount + interest
        print(f'💰 SBI Interest (6.5%): ₹{interest}')
        print(f'💵 Total Amount: ₹{total}')
        return total
    
    def withdrawal_limit(self):
        print('🏧 SBI Daily Withdrawal Limit: ₹40,000')
```

```python
# Customer chooses SBI Bank
print("📱 Customer 2: Using SBI Bank")
print("-"*60)
sbi_customer = SBIBank()
sbi_customer.customer_login('Priya')
sbi_customer.security()
sbi_customer.calculate_interest(100000, 2)
sbi_customer.withdrawal_limit()
```

```python
# Another Real-Time Example: Payment Gateway
from abc import ABC, abstractmethod

class PaymentGateway(ABC):
    
    # Common for all payment methods
    def verify_user(self, user_id):
        print(f'✅ User {user_id} verified')
    
    def send_receipt(self, email):
        print(f'📧 Receipt sent to {email}')
    
    # Must be implemented by each payment method
    @abstractmethod
    def process_payment(self, amount):
        pass
    
    @abstractmethod
    def payment_charges(self, amount):
        pass


class GooglePay(PaymentGateway):
    
    def process_payment(self, amount):
        print(f'💳 Processing ₹{amount} via Google Pay')
        print('✅ Payment Successful!')
    
    def payment_charges(self, amount):
        charges = amount * 0.01  # 1% charges
        print(f'💸 Google Pay Charges (1%): ₹{charges}')
        return charges


class PhonePe(PaymentGateway):
    
    def process_payment(self, amount):
        print(f'💳 Processing ₹{amount} via PhonePe')
        print('✅ Payment Successful!')
    
    def payment_charges(self, amount):
        charges = amount * 0.015  # 1.5% charges
        print(f'💸 PhonePe Charges (1.5%): ₹{charges}')
        return charges


class Paytm(PaymentGateway):
    
    def process_payment(self, amount):
        print(f'💳 Processing ₹{amount} via Paytm')
        print('✅ Payment Successful!')
    
    def payment_charges(self, amount):
        charges = amount * 0.02  # 2% charges
        print(f'💸 Paytm Charges (2%): ₹{charges}')
        return charges


# Customer makes payment
print('🛒 Customer ordering food for ₹500')
print('-'*50)

# Using Google Pay
gpay = GooglePay()
gpay.verify_user('USER123')
gpay.process_payment(500)
gpay.payment_charges(500)
gpay.send_receipt('user@gmail.com')

print('\n' + '-'*50)

# Using PhonePe
phonepe = PhonePe()
phonepe.verify_user('USER456')
phonepe.process_payment(500)
phonepe.payment_charges(500)
phonepe.send_receipt('user@gmail.com')
```

```python
# Simple Getter and Setter
class Person:
    def __init__(self, name):
        self._name = name

    @property
    def name(self):
        return self._name

    @name.setter
    def name(self, new_name):
        if not new_name:
            raise ValueError("Name cannot be empty")
        self._name = new_name

user = Person("Maya")
print(user.name)  # Output: Maya
user.name = "Sam"
print(user.name)  
```

```python
# @property decorator
class Temperature:
    def __init__(self, celsius):
        # Internal, "private" variable
        self._celsius = celsius

    # 1. Define the "getter" with @property
    @property
    def temp(self):
        print("Getting temperature...")
        return self._celsius
    
    # 2. Define the "setter" with @temp.setter
    @temp.setter
    def temp(self, value):
        if value < -273.15:
            raise ValueError("Temperature below absolute zero is not possible.")
        print("Setting temperature...")
        self._celsius = value
```

```python
# Usage
t = Temperature(25)
current_temp = t.temp
print(current_temp)

t.temp = 30 # Accesses the setter
# Outputs: Setting temperature...
t.temp
```
