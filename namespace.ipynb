````python
# local and global
# global scope
a = 2

def temp():
    # local scope
    b = 3
    print(b)

temp()
print(a)
````

````python
# local and global -> local does not have but global has
a = 2

def temp():
    print(a)

temp()
print(a)
````

````python
# local and global -> editing global
a = 2

def temp():
    # local var
    a += 1
    print(a)

temp()
print(a)
````

````python
a = 2

def temp():
    # local var
    global a
    a += 1
    print(a)

temp()
print(a)
````

````python
# local and global -> global created inside local
def temp():
    # local var
    global a 
    a = 1
    print(a)

temp()
print(a)
````

````python
# local and global -> function paramter is local
def temp(z):
    # local var 
    print(z)

a = 5 
temp(5)
print(a)
````

````python
# Enclosing Scope
def outer():
    x = 10   
    
    def inner():
        print(x)   
    
    inner()

outer()
````

````python
x = 100  # Global

def outer():
    x = 50   # Enclosing
    
    def inner():
        print(x)
    
    inner()

outer()
````

````python
# Modifying Enclosing Variable
def outer():
    x = 10
    
    def inner():
        x = 20   # New local variable
        print(x)
    
    inner()
    print(x)

outer()
````

````python
# To Modify Enclosing Variable → Use nonlocal
def outer():
    x = 10
    
    def inner():
        nonlocal x
        x = 20
        print("Inner:", x)
    
    inner()
    print("Outer:", x)

outer()
````

````python
# Real World Example (Counter System)
def counter():
    count = 0
    
    def increment():
        nonlocal count
        count += 1
        return count
    
    return increment

c = counter()

print(c())
print(c())
print(c())
````

````python
# Basic Decorator Structure
# Normal Func
def greet():
    print("Hello Zeeshan")

# Create Decorator
def my_decorator(func):
    
    def wrapper():
        print("Before Function")
        func()
        print("After Function")
    
    return wrapper

# Apply Decorator (Manual Way)
greet = my_decorator(greet)
greet()
````

````python
# Pythonic Way Using @ Symbol
@my_decorator
def greet():
    print("Hello Zeeshan")

greet()
````

````python
# anything meaningful?
import time 

def timer_dec(func):
    def wrapper(*args):
        start = time.time()
        func(*args)
        print(f'{func.__name__} function taken {time.time() - start} seconds to execute')
    return wrapper

@timer_dec
def hello():
    print("Hello World, welcome to Python Programming")
    time.sleep(2)

@timer_dec
def square(num):
    time.sleep(1)
    print(f"Square of {num} is: {num ** 2}")

@timer_dec
def power(a , b):
    print(f"Power of a and b is: {a ** b}")

hello()
print()
square(36)
print()
power(2 , 3)
````

````python
def sanity_check(data_type):
    def outer_wrapper(func):
        def inner_wrapper(*args):
            if type(*args) == data_type:
                func(*args)
            else:
                raise TypeError('Invalid Data Type')
        return inner_wrapper
    return outer_wrapper

@sanity_check(int)
def square(num):
    print(num ** 2)

@sanity_check(str)
def greet(name):
    print("Hello " , name)

square(4)
greet('Mike')
````

````python
# Decorator with Arguments (*args, **kwargs)
def smart_divide(func):
    def wrapper(a, b):
        print(f"🔢 Dividing {a} by {b}")
        if b == 0:
            print("❌ Error: Cannot divide by zero!")
            return None
        return func(a, b)
    return wrapper

@smart_divide
def divide(a, b):
    result = a / b
    print(f"✅ Result: {result}")
    return result
````

````python
# Test cases
divide(10, 2)
print()
divide(10, 0)
print()
divide(100, 5)
````

````python
# Best Practice (Use *args, **kwargs)
import time

def timer(func):
    
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()
        print("Execution Time:", end - start)
        return result
    
    return wrapper


@timer
def slow_function():
    time.sleep(2)
    print("Done")

slow_function()
````

````python
# Real-Time Example 1: Login Required Decorator
# Decorator to check if user is logged in
def login_required(func):
    def wrapper(user):
        if user.get('logged_in'):
            func(user)
        else:
            print("❌ Access Denied! Please login first.")
    return wrapper
````

````python
@login_required
def view_profile(user):
    print(f"✅ Welcome {user['name']}!")
    print(f"📧 Email: {user['email']}")
    print(f"🎂 Age: {user['age']}")


@login_required
def view_dashboard(user):
    print(f"📊 Dashboard for {user['name']}")
    print("📈 Your analytics are ready!")
````

````python
#  Real-Time Example 2: Timer Decorator (Measure Performance)
import time

def timer_decorator(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()
        print(f"⏰ Starting {func.__name__}...")
        
        result = func(*args, **kwargs)
        
        end_time = time.time()
        duration = end_time - start_time
        print(f"✅ {func.__name__} completed in {duration:.4f} seconds")
        return result
    return wrapper
````

````python
@timer_decorator
def download_file():
    print("📥 Downloading file...")
    time.sleep(2)  # Simulate download time
    print("✅ Download complete!")


@timer_decorator
def process_data():
    print("⚙️ Processing data...")
    time.sleep(1.5)
    print("✅ Processing complete!")


@timer_decorator
def calculate_sum(n):
    print(f"🔢 Calculating sum of first {n} numbers...")
    total = sum(range(1, n+1))
    print(f"💯 Sum = {total}")
    return total
````

````python
# Using the decorated functions
download_file()
print()
process_data()
print()
calculate_sum(1000000)
````

````python
# Multiple Decorators (Stacking)
def bold_decorator(func):
    def wrapper():
        return "<b>" + func() + "</b>"
    return wrapper

def italic_decorator(func):
    def wrapper():
        return "<i>" + func() + "</i>"
    return wrapper

def underline_decorator(func):
    def wrapper():
        return "<u>" + func() + "</u>"
    return wrapper


# Apply multiple decorators
@bold_decorator
@italic_decorator
@underline_decorator
def greet():
    return "Hello World"


print(greet())
# Output: <b><i><u>Hello World</u></i></b>
# ```

# **Order of execution (bottom to top):**
# ```
# 1. underline_decorator wraps greet() → <u>Hello World</u>
# 2. italic_decorator wraps that → <i><u>Hello World</u></i>
# 3. bold_decorator wraps that → <b><i><u>Hello World</u></i></b>
````

````python
# Create Sanity Check Decorator
#  Example 1: Type Checking Decorator (@sanity_check)
# Decorator with parameter to check data types
def sanity_check(expected_type):
    """Outer function - receives the parameter (int, str, float, etc.)"""
    
    def decorator(func):
        """Middle function - receives the actual function"""
        
        def wrapper(*args, **kwargs):
            """Inner function - does the actual work"""
            
            # Check all arguments
            for arg in args:
                if not isinstance(arg, expected_type):
                    print(f"❌ Type Error: Expected {expected_type.__name__}, got {type(arg).__name__}")
                    return None
            
            # If all checks pass, run the function
            print(f"✅ Type check passed! All arguments are {expected_type.__name__}")
            return func(*args, **kwargs)
        
        return wrapper
    return decorator


# Test with integers
@sanity_check(int)
def add_numbers(a, b, c):
    result = a + b + c
    print(f"🔢 Sum: {result}")
    return result


# Test with strings
@sanity_check(str)
def join_strings(s1, s2, s3):
    result = s1 + " " + s2 + " " + s3
    print(f"📝 Joined: {result}")
    return result


# Test with floats
@sanity_check(float)
def calculate_average(x, y, z):
    result = (x + y + z) / 3
    print(f"📊 Average: {result:.2f}")
    return result


# ==================== TESTING ====================
print("="*60)
print("Test 1: Integer function with correct types")
print("-"*60)
add_numbers(10, 20, 30)

print("\n" + "="*60)
print("Test 2: Integer function with wrong types")
print("-"*60)
add_numbers(10, "hello", 30)  # String passed!

print("\n" + "="*60)
print("Test 3: String function with correct types")
print("-"*60)
join_strings("Hello", "Python", "World")

print("\n" + "="*60)
print("Test 4: String function with wrong types")
print("-"*60)
join_strings("Hello", 123, "World")  # Integer passed!

print("\n" + "="*60)
print("Test 5: Float function with correct types")
print("-"*60)
calculate_average(10.5, 20.3, 15.7)

print("\n" + "="*60)
print("Test 6: Float function with wrong types")
print("-"*60)
calculate_average(10.5, 20, 15.7)  # Integer passed instead of float!
````
