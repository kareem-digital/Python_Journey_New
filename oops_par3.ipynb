````python
# Aggregation (Has-A)
class Customer:
    def __init__(self , name , gender , address):
        self.name = name
        self.gender = gender
        self.address = address

    def display(self):
        print(f"Name : {self.name}")
        print(f"City : {self.address.get_city()}")
        print(f"Pin : {self.address.pin}")
        print(f"State : {self.address.state}")

    def edit_profile(self , new_name, new_city, new_pin, new_state):
        self.name = new_name
        self.address.edit_address(new_city, new_pin , new_state)


class Address:
    def __init__(self , city , pin, state):
        self.__city = city
        self.pin = pin
        self.state = state

    def get_city(self):
        return self.__city
    
    def edit_address(self, new_city , new_pin, new_state):
        self.__city = new_city
        self.pin = new_pin
        self.state = new_state

add1 = Address('Mau' , 276129 , 'UP')
cust = Customer('Rahul' , 'Male' , add1)
cust.display()
print("="*40)

cust.edit_profile('Mohan', 'Delhi' , 226001 , 'Delhi')
cust.display()
````

````python
# ════════════════════════════════════
# AGGREGATION EXAMPLE
# ════════════════════════════════════

class Employee:
    def __init__(self, name, emp_id):
        self.name = name
        self.emp_id = emp_id
    
    def display(self):
        print(f"  • {self.name} ({self.emp_id})")

class Department:
    def __init__(self, dept_name):
        self.dept_name = dept_name
        self.employees = []  # List of Employee objects
    
    def add_employee(self, employee):
        """Add existing employee to department"""
        self.employees.append(employee)
        print(f"✅ {employee.name} added to {self.dept_name}")
    
    def remove_employee(self, employee):
        """Remove employee from department"""
        if employee in self.employees:
            self.employees.remove(employee)
            print(f"✅ {employee.name} removed from {self.dept_name}")
    
    def show_employees(self):
        print(f"\n{self.dept_name} Department Employees:")
        if not self.employees:
            print("  (No employees)")
        else:
            for emp in self.employees:
                emp.display()
````

````python
# ==========================================
# TESTING AGGREGATION
# ==========================================

print("="*60)
print("AGGREGATION (Weak HAS-A Relationship)")
print("="*60)

# Create employees (exist independently)
emp1 = Employee("Rahul Sharma", "E001")
emp2 = Employee("Priya Singh", "E002")
emp3 = Employee("Amit Kumar", "E003")

# Create departments
it_dept = Department("IT")
hr_dept = Department("HR")

# Add employees to departments
print("\n--- Adding Employees to Departments ---")
it_dept.add_employee(emp1)
it_dept.add_employee(emp2)
hr_dept.add_employee(emp3)

# Show employees
it_dept.show_employees()
hr_dept.show_employees()

# Employee can exist without department
print("\n--- Removing Employee from Department ---")
it_dept.remove_employee(emp1)
it_dept.show_employees()

# emp1 still exists even after removal!
print("\n--- Employee still exists independently ---")
emp1.display()
````

````python
# Composition
class Heart:
    def __init__(self, heartValves):
        self.heartValves = heartValves
        
    def display(self):
        return self.heartValves
    
class Person:
    def __init__(self, fname, lname, address, heartValves):
        self.fname = fname
        self.lname = lname
        self.address = address
        self.heartValves = heartValves
        self.heartObject = Heart(self.heartValves)   # Composition
        
    def display(self):
        print("First Name: ", self.fname)
        print("Last Name: ", self.lname)
        print("Address: ", self.address)
        print("No of Heart Valves: ", self.heartObject.display())


p = Person("Adam", "syn", "876 Zyx Ln", 4)
p.display()
````

````python
# Code to demonstrate Composition
# Class Salary in which we are
# declaring a public method annual salary
class Salary:
    def __init__(self, pay, bonus):
        self.pay = pay
        self.bonus = bonus

    def annual_salary(self):
        return (self.pay*12)+self.bonus


# Class EmployeeOne which does not 
# inherit the class Salary yet we will
# use the method annual salary using
# Composition
class EmployeeOne:
    def __init__(self, name, age, pay, bonus):
        self.name = name
        self.age = age

        # Making an object in which we are
        # calling the Salary class
        # with proper arguments.
        self.obj_salary = Salary(pay, bonus)  # composition


    # Method which calculates the total salary
    # with the help of annual_salary() method
    # declared in the Salary class
    def total_sal(self):
        return self.obj_salary.annual_salary()
    
# Making an object of the class EmployeeOne
# and providing necessary arguments
emp = EmployeeOne('Geek', 25, 10000, 1500)
# calling the total_sal method using 
# the emp object
print(emp.total_sal())
````

````python
# ════════════════════════════════════
# COMPOSITION EXAMPLE
# ════════════════════════════════════
class Engine:
    def __init__(self, capacity, engine_type):
        self.capacity = capacity
        self.engine_type = engine_type
        print(f"  🔧 Engine created: {capacity}L {engine_type}")
    
    def start(self):
        print(f"  🔧 {self.capacity}L engine starting...")
    
    def __del__(self):
        print(f"  🔧 Engine destroyed")

class Wheel:
    def __init__(self, size):
        self.size = size
        print(f"  🛞 Wheel created: {size} inches")
    
    def __del__(self):
        print(f"  🛞 Wheel destroyed")

class Car:
    def __init__(self, brand, model, engine_capacity):
        self.brand = brand
        self.model = model
        
        # Composition: Car creates its own Engine
        # Engine cannot exist without Car
        self.engine = Engine(engine_capacity, "Petrol")
        
        # Composition: Car creates its own Wheels
        self.wheels = [Wheel(17) for _ in range(4)]
        
        print(f"✅ Car created: {brand} {model}")
    
    def start(self):
        print(f"\n🚗 Starting {self.brand} {self.model}...")
        self.engine.start()
        print(f"  🛞 All {len(self.wheels)} wheels ready")
        print(f"🚗 {self.brand} {self.model} is running!")
    
    def __del__(self):
        print(f"\n🚗 Car {self.brand} {self.model} destroyed")
        # Engine and Wheels are automatically destroyed
````

````python
# ==========================================
# TESTING COMPOSITION
# ==========================================

print("="*60)
print("COMPOSITION (Strong HAS-A Relationship)")
print("="*60)

print("\n--- Creating Car (with Engine & Wheels) ---")
car1 = Car("Toyota", "Camry", 2.5)

print("\n--- Using Car ---")
car1.start()

print("\n--- Deleting Car ---")
print("When car is deleted, engine and wheels are also deleted:")
del car1
# Engine and Wheels are destroyed along with Car!

print("\n" + "="*60)
````

````python
# 1. Single Inheritance
# One parent → One child
class Phone:
    def __init__(self, brand, price):
        self.brand = brand
        self.price = price

    def call(self):
        print(f"{self.brand} is making a call")

class SmartPhone(Phone):              # Single parent
    def __init__(self, brand, price, os):
        super().__init__(brand, price)
        self.os = os

    def browse_internet(self):
        print(f"Browsing internet on {self.brand}")

s = SmartPhone("Apple", 200000, "iOS")
s.call()              # ✅ Inherited from Phone
s.browse_internet()   # ✅ Own method
print(s.os)           # ✅ iOS
````

````python
## 2. Multilevel Inheritance
# **One parent → Child → Grandchild (chain)**
# ```
# Phone
#   ↓
# SmartPhone
#   ↓
# iPhone
class Phone:
    def __init__(self, brand, price):
        self.brand = brand
        self.price = price

    def call(self):
        print(f"{self.brand} is making a call")

class SmartPhone(Phone):
    def __init__(self, brand, price, os):
        super().__init__(brand, price)
        self.os = os

    def browse_internet(self):
        print(f"Browsing internet on {self.brand}")

class iPhone(SmartPhone):              # Inherits from SmartPhone
    def __init__(self, brand, price, os, chip):
        super().__init__(brand, price, os)
        self.chip = chip

    def face_id(self):
        print(f"Face ID unlocked on {self.brand} with {self.chip} chip")

i = iPhone("Apple", 200000, "iOS", "A17 Pro")

i.call()              # ✅ From Phone (Grandparent)
i.browse_internet()   # ✅ From SmartPhone (Parent)
i.face_id()           # ✅ Own method

print(i.brand)        # ✅ Apple
print(i.os)           # ✅ iOS
print(i.chip)         # ✅ A17 Pro
````

````python
## 3. Multiple Inheritance
# **Two or more parents → One child**
class Phone:
    def __init__(self, brand, price):
        self.brand = brand
        self.price = price

    def call(self):
        print(f"{self.brand} is making a call")

class Camera:
    def __init__(self, megapixels):
        self.megapixels = megapixels

    def click_photo(self):
        print(f"Clicking photo with {self.megapixels}MP camera")

class SmartPhone(Phone, Camera):       # Two parents!
    def __init__(self, brand, price, megapixels, os):
        Phone.__init__(self, brand, price)       # Parent 1
        Camera.__init__(self, megapixels)        # Parent 2
        self.os = os

    def browse_internet(self):
        print(f"Browsing internet on {self.brand}")

s = SmartPhone("Samsung", 150000, 200, "Android")

s.call()              # ✅ From Phone
s.click_photo()       # ✅ From Camera
s.browse_internet()   # ✅ Own method

print(s.brand)        # ✅ Samsung
print(s.megapixels)   # ✅ 200
print(s.os)           # ✅ Android
````

````python
## 4. Hierarchical Inheritance
# **One parent → Multiple children**
class Phone:
    def __init__(self, brand, price):
        self.brand = brand
        self.price = price

    def call(self):
        print(f"{self.brand} is making a call")

class SmartPhone(Phone):               # Child 1
    def __init__(self, brand, price, os):
        super().__init__(brand, price)
        self.os = os

    def browse_internet(self):
        print(f"Browsing internet on {self.brand} running {self.os}")

class FeaturePhone(Phone):             # Child 2
    def __init__(self, brand, price, battery):
        super().__init__(brand, price)
        self.battery = battery

    def long_battery(self):
        print(f"{self.brand} has {self.battery}mAh long battery life")

s = SmartPhone("Apple", 200000, "iOS")
s.call()              # ✅ From Phone
s.browse_internet()   # ✅ Own method

f = FeaturePhone("Nokia", 3000, 7000)
f.call()              # ✅ From Phone
f.long_battery()      # ✅ Own method
````

````python
## 5. Hybrid Inheritance
# **Mix of multiple types**
class Phone:
    def __init__(self, brand, price):
        self.brand = brand
        self.price = price

    def call(self):
        print(f"{self.brand} is making a call")

class SmartPhone(Phone):
    def __init__(self, brand, price, os):
        super().__init__(brand, price)
        self.os = os

    def browse_internet(self):
        print(f"Browsing on {self.brand}")

class Camera(Phone):
    def __init__(self, brand, price, megapixels):
        super().__init__(brand, price)
        self.megapixels = megapixels

    def click_photo(self):
        print(f"Clicking {self.megapixels}MP photo")

class iPhone(SmartPhone, Camera):      # Hybrid!
    def __init__(self, brand, price, os, megapixels, chip):
        SmartPhone.__init__(self, brand, price, os)
        Camera.__init__(self, brand, price, megapixels)
        self.chip = chip

    def face_id(self):
        print(f"Face ID on {self.brand} with {self.chip}")

i = iPhone("Apple", 200000, "iOS", 48, "A17 Pro")

i.call()              # ✅ From Phone
i.browse_internet()   # ✅ From SmartPhone
i.click_photo()       # ✅ From Camera
i.face_id()           # ✅ Own method

print(i.brand)        # ✅ Apple
print(i.os)           # ✅ iOS
print(i.megapixels)   # ✅ 48
print(i.chip)         # ✅ A17 Pro
````

````python
class Phone:
    def __init__(self, brand, price, **kwargs):
        self.brand = brand
        self.price = price

    def call(self):
        print(f"{self.brand} is making a call")

class SmartPhone(Phone):
    def __init__(self, brand, price, os, **kwargs):
        super().__init__(brand, price, **kwargs)
        self.os = os

    def browse_internet(self):
        print(f"Browsing on {self.brand}")

class Camera(Phone):
    def __init__(self, brand, price, megapixels, **kwargs):
        super().__init__(brand, price, **kwargs)
        self.megapixels = megapixels

    def click_photo(self):
        print(f"Clicking {self.megapixels}MP photo")

class iPhone(SmartPhone, Camera):
    def __init__(self, brand, price, os, megapixels, chip):
        super().__init__(brand=brand, price=price, os=os, megapixels=megapixels)
        self.chip = chip

    def face_id(self):
        print(f"Face ID on {self.brand} with {self.chip}")

i = iPhone("Apple", 200000, "iOS", 48, "A17 Pro")

i.call()             # ✅ From Phone
i.browse_internet()  # ✅ From SmartPhone
i.click_photo()      # ✅ From Camera
i.face_id()          # ✅ Own method

print(i.brand)       # ✅ Apple
print(i.os)          # ✅ iOS
print(i.megapixels)  # ✅ 48
print(i.chip)        # ✅ A17 Pro
````

````python
# The simplest form. One class inherits from one parent.
class Animal:
    def __init__(self, name):
        self.name = name

    def eat(self):
        print(f"{self.name} is eating")

    def sleep(self):
        print(f"{self.name} is sleeping")

class Dog(Animal):                  # Dog inherits from Animal
    def __init__(self, name):
        super().__init__(name)      # Get name from Animal

    def bark(self):
        print(f"{self.name} says Woof!")

d = Dog("Bruno")

d.eat()     # ✅ Inherited from Animal → Bruno is eating
d.sleep()   # ✅ Inherited from Animal → Bruno is sleeping
d.bark()    # ✅ Own method           → Bruno says Woof!

# **Real Life Analogy:**
# > A son inherits his father's property + has his own car.

````

````python
## 2. Multilevel Inheritance
### **Parent → Child → Grandchild (A Chain)**
# Like a grandfather → father → son chain.
class Animal:
    def __init__(self, name):
        self.name = name

    def eat(self):
        print(f"{self.name} is eating")

class Dog(Animal):                  # Dog inherits Animal
    def __init__(self, name):
        super().__init__(name)

    def bark(self):
        print(f"{self.name} says Woof!")

class Puppy(Dog):                   # Puppy inherits Dog
    def __init__(self, name):
        super().__init__(name)

    def play(self):
        print(f"{self.name} is playing!")

p = Puppy("Tommy")

p.eat()    # ✅ From Animal (Grandparent) → Tommy is eating
p.bark()   # ✅ From Dog (Parent)         → Tommy says Woof!
p.play()   # ✅ Own method                → Tommy is playing!

# **Real Life Analogy:**
# > Grandfather owns land → Father inherits land + builds house → Son inherits land + house + buys a car.

````

````python
## 3. Multiple Inheritance
### **Two or More Parents → One Child**
# A child inherits from two different parents at the same time.
class Father:
    def __init__(self):
        self.father_name = "Rajesh"

    def coding(self):
        print(f"{self.father_name} is good at coding")

class Mother:
    def __init__(self):
        self.mother_name = "Sunita"

    def cooking(self):
        print(f"{self.mother_name} is good at cooking")

class Child(Father, Mother):        # Two parents!
    def __init__(self):
        Father.__init__(self)
        Mother.__init__(self)

    def playing(self):
        print("Child loves playing!")

c = Child()

c.coding()    # ✅ From Father → Rajesh is good at coding
c.cooking()   # ✅ From Mother → Sunita is good at cooking
c.playing()   # ✅ Own method  → Child loves playing!

# **Real Life Analogy:**
# > A child gets dad's height and mom's eyes — traits from BOTH parents.
````

````python
## 4. Hierarchical Inheritance
### **One Parent → Multiple Children**
# One parent, but many children each with their own features.
class Animal:
    def __init__(self, name):
        self.name = name

    def eat(self):
        print(f"{self.name} is eating")

    def sleep(self):
        print(f"{self.name} is sleeping")

class Dog(Animal):                  # Child 1
    def __init__(self, name):
        super().__init__(name)

    def bark(self):
        print(f"{self.name} says Woof!")

class Cat(Animal):                  # Child 2
    def __init__(self, name):
        super().__init__(name)

    def meow(self):
        print(f"{self.name} says Meow!")

class Cow(Animal):                  # Child 3
    def __init__(self, name):
        super().__init__(name)

    def moo(self):
        print(f"{self.name} says Moo!")

d = Dog("Bruno")
d.eat()    # ✅ From Animal
d.bark()   # ✅ Own method

c = Cat("Kitty")
c.eat()    # ✅ From Animal (same parent!)
c.meow()   # ✅ Own method

cow = Cow("Gowri")
cow.eat()  # ✅ From Animal (same parent!)
cow.moo()  # ✅ Own method


# **Real Life Analogy:**
# > One father has 3 children. All 3 inherit the same property. But each child has their own job — engineer, doctor, teacher.

````

````python
## 5. Hybrid Inheritance
### **Mix of 2 or more types above**
# A combination of multiple inheritance types together.
class Animal:
    def __init__(self, name):
        self.name = name

    def eat(self):
        print(f"{self.name} is eating")

class Dog(Animal):                  # Hierarchical
    def __init__(self, name):
        super().__init__(name)

    def bark(self):
        print(f"{self.name} says Woof!")

class Cat(Animal):                  # Hierarchical
    def __init__(self, name):
        super().__init__(name)

    def meow(self):
        print(f"{self.name} says Meow!")

class HybridPet(Dog, Cat):          # Multiple
    def __init__(self, name):
        super().__init__(name)

    def tricks(self):
        print(f"{self.name} can do tricks!")

h = HybridPet("Buddy")

h.eat()     # ✅ From Animal → Buddy is eating
h.bark()    # ✅ From Dog    → Buddy says Woof!
h.meow()    # ✅ From Cat    → Buddy says Meow!
h.tricks()  # ✅ Own method  → Buddy can do tricks!


# **Real Life Analogy:**
# > A person who is both a doctor AND an engineer — inherits skills from two different fields.





````
