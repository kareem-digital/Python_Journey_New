```python
# Using getattr() with Default Value
class Student:
    def __init__(self, name):
        self.name = name

student1 = Student("Rahul")
student1.age = 16

student2 = Student("Priya")

# getattr(object, attribute, default_value)
age1 = getattr(student1, 'age', 'Not specified')
print(age1)  # 16

age2 = getattr(student2, 'age', 'Not specified')
print(age2)  # Not specified (doesn't have age)

# Also works for any attribute
grade1 = getattr(student1, 'grade', 'Unknown')
print(grade1)  # Unknown
```

```python
# ATM
class Atm:
    # constructor
    def __init__(self):
        self.pin = ''
        self.balance = 0 
        # self.menu()

    # atm menu
    def menu(self):
        user_input = input("""
            Hi how i can help you?
            1. Press 1 to create pin
            2. Press 2 to change pin
            3. Press 3 check balance
            4. Press 4 to withdraw
            5. Anything else to exit 
        """)

        if user_input == "1":
            # create pin
            self.create_pin()
        elif user_input == "2":
            # change pin
            self.change_pin()
        elif user_input == "3":
            # check balance 
            self.check_balane()
        elif user_input == "4":
            # withdraw
            self.withdraw()
        else:
            exit()

    def create_pin(self):
        user_pin = input("Enter your pin")
        self.pin = user_pin

        user_balance = int(input("Enter your balance"))
        self.balance = user_balance

        print('pin created successfully')
        self.menu()

    def change_pin(self):
        old_pin = input("Enter old pin")
        if old_pin == self.pin:
            # let him change the pin 
            new_pin = input("Enter new pin")
            self.pin = new_pin 
            print("pin change successfully!")
            self.menu()
        else:
            print("Not possible! at this time")
            self.menu()

    def check_balane(self):
        user_pin = input("enter your pin")
        if user_pin == self.pin:
            print(f"Your balance is {self.balance}")
        else:
            print("Wrong pin, Try again")
        

    def withdraw(self):
        user_pin = input("enter your pin")
        if user_pin == self.pin:
            # allow to withdraw
            amount = int(input("enter the amount"))
            if amount <= self.balance:
                self.balance = self.balance - amount
                print('Withdrawl Successful')
                print(f"Remaining Balance: {self.balance}")
            else:
                print("Insufficient Balance")
        else:
            print("Wrong pin, Try again")
        self.menu()
```

##### Getters and Setters in Python OOP

```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance  # Private variable
    
    # GETTER - Get balance
    def get_balance(self):
        print("🔍 Balance checked")  # Can add logging
        return self.__balance
    
    # SETTER - Set balance
    def set_balance(self, amount):
        # VALIDATION before setting
        if amount < 0:
            print("❌ Cannot set negative balance!")
            return
        
        if not isinstance(amount, (int, float)):
            print("❌ Balance must be a integer!")
            return
        
        print(f"✅ Balance updated to ₹{amount}")
        self.__balance = amount

account = BankAccount(10000)
```

```python
# Another examples
class Student:
    def __init__(self, name, age, marks):
        # Private variables
        self.__name = name
        self.__age = age
        self.__marks = marks

    # ═══════════════════════════════════
    # GETTERS (Read values)
    # ═══════════════════════════════════
    
    def get_name(self):
        """Get student name"""
        return self.__name
    
    def get_age(self):
        """Get student age"""
        return self.__age
    
    def get_marks(self):
        """Get student marks"""
        return self.__marks

    # ═══════════════════════════════════
    # SETTERS (Write values with validation)
    # ═══════════════════════════════════
    
    def set_name(self, name):
        """Set student name"""
        if len(name.strip()) == 0:
            print("❌ Name cannot be empty!")
            return
        
        if not name.replace(" ", "").isalpha():
            print("❌ Name must contain only letters!")
            return
        
        self.__name = name
        print(f"✅ Name updated to: {name}")
    
    def set_age(self, age):
        """Set student age with validation"""
        if not isinstance(age, int):
            print("❌ Age must be an integer!")
            return
        
        if age < 5 or age > 100:
            print("❌ Age must be between 5 and 100!")
            return
        
        self.__age = age
        print(f"✅ Age updated to: {age}")
    
    def set_marks(self, marks):
        """Set student marks with validation"""
        if not isinstance(marks, (int, float)):
            print("❌ Marks must be a number!")
            return
        
        if marks < 0 or marks > 100:
            print("❌ Marks must be between 0 and 100!")
            return
        
        self.__marks = marks
        print(f"✅ Marks updated to: {marks}")
    
    def display(self):
        """Display student information"""
        print(f"\n{'='*40}")
        print(f"Name: {self.__name}")
        print(f"Age: {self.__age}")
        print(f"Marks: {self.__marks}")
        print(f"{'='*40}")

# ==========================================
# USING GETTERS AND SETTERS
# ==========================================

student = Student("Rahul Sharma", 16, 85)
```

##### The @property Decorator (Pythonic Way) - create getters and setters

```python
class Student:
    def __init__(self, name, age):
        self.__name = name
        self.__age = age
    
    # ═══ GETTER using @property ═══
    @property
    def name(self):
        """Get name"""
        return self.__name
    
    @property
    def age(self):
        """Get age"""
        return self.__age
    
    # ═══ SETTER using @<property>.setter ═══
    @name.setter
    def name(self, value):
        """Set name with validation"""
        if len(value.strip()) == 0:
            raise ValueError("Name cannot be empty!")
        self.__name = value
    
    @age.setter
    def age(self, value):
        """Set age with validation"""
        if not 5 <= value <= 100:
            raise ValueError("Age must be between 5 and 100!")
        self.__age = value
    
    def display(self):
        print(f"Name: {self.__name}, Age: {self.__age}")

# ==========================================
# USING @property (Looks like normal attributes!)
# ==========================================

student = Student("Rahul", 16)
```

```python
# Access like normal attributes (but uses getter internally)
print(student.name)  # Rahul
print(student.age)   # 16
```

```python
# Set like normal attributes (but uses setter with validation)
student.name = "Priya"  # ✅ Valid
student.age = 17        # ✅ Valid
```

```python
print(student.name)  # Priya
print(student.age)   # 17
```

```python
# Validation still happens!
try:
    student.name = ""    # ❌ Raises ValueError
except ValueError as e:
    print(f"Error: {e}")
```

```python
try:
    student.age = 150    # ❌ Raises ValueError
except ValueError as e:
    print(f"Error: {e}")
```

```python
class Celsius:
    def __init__(self, temperature=0):
        self._temperature = temperature

    @property
    def temperature(self):
        print("Getting value...")
        return self._temperature

    @temperature.setter
    def temperature(self, value):
        if value < -273.15:
            raise ValueError("Temperature below -273.15 is not possible")
        print("Setting value...")
        self._temperature = value

# Usage
obj = Celsius()
obj.temperature = 30  # Calls setter
print(obj.temperature) # Calls getter (no parentheses needed)
```

##### Static Method:

```python
class Employee:
    # Static variable
    company = "TechCorp"
    total_employees = 0
    
    def __init__(self, name, salary):
        # Instance variables
        self.name = name
        self.salary = salary
        Employee.total_employees += 1
    
    # ════════════════════════════════════
    # 1. INSTANCE METHOD
    # Uses self - accesses instance data
    # ════════════════════════════════════
    def display_info(self):
        print(f"Name: {self.name}")      # instance var
        print(f"Salary: ₹{self.salary}") # instance var
        print(f"Company: {Employee.company}")  # class var
    
    def give_raise(self, amount):
        self.salary += amount  # Modifies instance var
        print(f"✅ {self.name}'s salary raised by ₹{amount}")
    
    # ════════════════════════════════════
    # 2. CLASS METHOD
    # Uses cls - accesses class data
    # ════════════════════════════════════
    @classmethod
    def change_company(cls, new_company):
        cls.company = new_company  # Modifies class var
        print(f"✅ Company changed to: {new_company}")
    
    @classmethod
    def get_total_employees(cls):
        return cls.total_employees  # Reads class var
    
    # ════════════════════════════════════
    # 3. STATIC METHOD
    # No self or cls - utility function
    # ════════════════════════════════════
    @staticmethod
    def validate_salary(salary):
        """Validates salary without needing object"""
        return salary > 0 and salary <= 10000000
    
    @staticmethod
    def calculate_tax(salary):
        """Calculates tax - doesn't need object data"""
        if salary <= 300000:
            return 0
        elif salary <= 600000:
            return salary * 0.05
        elif salary <= 900000:
            return salary * 0.10
        else:
            return salary * 0.15
    
    @staticmethod
    def is_valid_name(name):
        """Validates name format"""
        return len(name) >= 2 and name.replace(" ", "").isalpha()
```

```python
# ==========================================
# USING ALL THREE METHODS
# ==========================================

# Using STATIC METHOD before creating object!
print("=== Static Methods (No object needed!) ===")
print(f"Valid salary? {Employee.validate_salary(50000)}")   # True
print(f"Valid salary? {Employee.validate_salary(-100)}")    # False
print(f"Valid name? {Employee.is_valid_name('Rahul')}")     # True
print(f"Valid name? {Employee.is_valid_name('R')}")         # False
print(f"Tax on ₹700000: ₹{Employee.calculate_tax(700000)}")  # 70000.0
```

```python
# Create employees
e1 = Employee("Rahul Sharma", 75000)
e2 = Employee("Priya Singh", 65000)

print("\n=== Instance Methods ===")
e1.display_info()
e1.give_raise(5000)
e1.display_info()
```

##### Static Method vs Instance Method vs Class Method

```python
class Student:
    # ════════════════════════════════════
    # CLASS VARIABLE (Static)
    # ════════════════════════════════════
    school_name = "Delhi Public School"
    total_students = 0
    passing_marks = 35
    
    def __init__(self, name, age, marks):
        # ════════════════════════════════════
        # INSTANCE VARIABLES
        # ════════════════════════════════════
        self.name = name
        self.age = age
        self.marks = marks
        Student.total_students += 1
    
    # ════════════════════════════════════
    # 1. INSTANCE METHOD
    # ════════════════════════════════════
    # ✅ Has 'self' parameter
    # ✅ Can access instance variables
    # ✅ Can access class variables
    # ✅ Called on object: obj.method()
    
    def display_info(self):
        """Display student information"""
        print(f"\n{'='*45}")
        print(f"Name: {self.name}")           # Instance var
        print(f"Age: {self.age}")             # Instance var
        print(f"Marks: {self.marks}")         # Instance var
        print(f"School: {Student.school_name}") # Class var
        print(f"{'='*45}")
    
    def check_result(self):
        """Check if student passed"""
        if self.marks >= Student.passing_marks:  # Uses both
            return "PASSED ✅"
        return "FAILED ❌"
    
    def give_bonus(self, bonus_marks):
        """Add bonus marks to this student"""
        self.marks += bonus_marks  # Modifies instance var
        print(f"✅ Added {bonus_marks} bonus marks to {self.name}")
    
    # ════════════════════════════════════
    # 2. CLASS METHOD
    # ════════════════════════════════════
    # ✅ Has @classmethod decorator
    # ✅ Has 'cls' parameter
    # ✅ Can access class variables
    # ❌ Cannot access instance variables
    # ✅ Called on class: Class.method()
    
    @classmethod
    def change_school(cls, new_school):
        """Change school name for ALL students"""
        cls.school_name = new_school  # Modifies class var
        print(f"✅ School changed to: {new_school}")
    
    @classmethod
    def get_total_students(cls):
        """Get total number of students"""
        return cls.total_students  # Returns class var
    
    @classmethod
    def update_passing_marks(cls, new_marks):
        """Update passing marks for ALL students"""
        cls.passing_marks = new_marks
        print(f"✅ Passing marks updated to: {new_marks}")
    
    @classmethod
    def from_string(cls, student_string):
        """Factory method: Create student from string"""
        name, age, marks = student_string.split(",")
        return cls(name, int(age), int(marks))  # Create new object
    
    # ════════════════════════════════════
    # 3. STATIC METHOD
    # ════════════════════════════════════
    # ✅ Has @staticmethod decorator
    # ❌ No 'self' parameter
    # ❌ No 'cls' parameter
    # ❌ Cannot access instance variables
    # ❌ Cannot access class variables (directly)
    # ✅ Called on class: Class.method()
    # ✅ Just a utility function
    
    @staticmethod
    def validate_name(name):
        """Validate student name"""
        return len(name) >= 2 and name.replace(" ", "").isalpha()
    
    @staticmethod
    def validate_age(age):
        """Validate student age"""
        return 5 <= age <= 100
    
    @staticmethod
    def validate_marks(marks):
        """Validate marks range"""
        return 0 <= marks <= 100
    
    @staticmethod
    def calculate_percentage(marks, total_marks=100):
        """Calculate percentage"""
        return (marks / total_marks) * 100
    
    @staticmethod
    def grade_from_marks(marks):
        """Get grade based on marks"""
        if marks >= 90:
            return 'A+'
        elif marks >= 80:
            return 'A'
        elif marks >= 70:
            return 'B'
        elif marks >= 60:
            return 'C'
        else:
            return 'F'

# ==========================================
# TESTING ALL THREE METHODS
# ==========================================

print("="*60)
print("DEMONSTRATING ALL THREE METHOD TYPES")
print("="*60)
```

```python
# ═══════════════════════════════════════
# STATIC METHODS (No object needed!)
# ═══════════════════════════════════════
print("\n1. STATIC METHODS (Called before creating objects!)")
print("-" * 60)

# Can call without ANY object!
print(f"Valid name 'Rahul'? {Student.validate_name('Rahul')}")      # True
print(f"Valid name 'R'? {Student.validate_name('R')}")              # False
print(f"Valid age 16? {Student.validate_age(16)}")                  # True
print(f"Valid age 150? {Student.validate_age(150)}")                # False
print(f"Valid marks 85? {Student.validate_marks(85)}")              # True
print(f"Percentage of 85/100: {Student.calculate_percentage(85)}%")
print(f"Grade for 85 marks: {Student.grade_from_marks(85)}")
```

```python
# ═══════════════════════════════════════
# CLASS METHODS
# ═══════════════════════════════════════
print("\n2. CLASS METHODS (Work with class data)")
print("-" * 60)

# Create students normally
s1 = Student("Rahul Sharma", 16, 85)
s2 = Student("Priya Singh", 17, 92)

# Class method - get total students
print(f"Total students: {Student.get_total_students()}")  # 2

# Class method - change school (affects ALL students)
Student.change_school("Kendriya Vidyalaya")

# Class method - factory method to create object
s3 = Student.from_string("Amit Kumar,15,78")
print(f"✅ Created student from string: {s3.name}")

print(f"Total students now: {Student.get_total_students()}")
```

```python
# ═══════════════════════════════════════
# INSTANCE METHODS
# ═══════════════════════════════════════
print("\n3. INSTANCE METHODS (Work with object data)")
print("-" * 60)

# Must call on an object
s1.display_info()
print(f"Result: {s1.check_result()}")
s1.give_bonus(5)
print(f"New marks: {s1.marks}")

s2.display_info()
print(f"Result: {s2.check_result()}")

s3.display_info()
print(f"Result: {s3.check_result()}")

# ═══════════════════════════════════════
# SHOWING THE DIFFERENCES
# ═══════════════════════════════════════
print("\n" + "="*60)
print("DEMONSTRATING KEY DIFFERENCES")
print("="*60)

print("\n✅ Class Method - affects ALL objects:")
Student.update_passing_marks(40)
print(f"s1 result: {s1.check_result()}")  # All students affected
print(f"s2 result: {s2.check_result()}")
print(f"s3 result: {s3.check_result()}")  # Now fails (78 < 40? No, but shows it works)

print("\n✅ Instance Method - affects only ONE object:")
s1.give_bonus(10)
print(f"s1 marks: {s1.marks}")  # Only s1 changed
print(f"s2 marks: {s2.marks}")  # s2 unchanged
print(f"s3 marks: {s3.marks}")  # s3 unchanged

print("\n✅ Static Method - independent utility:")
# Works with ANY data, not tied to objects or class
for marks in [45, 65, 85, 95]:
    grade = Student.grade_from_marks(marks)
    print(f"Marks: {marks} → Grade: {grade}")
```
